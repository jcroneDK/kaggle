===================================================SCALING=================================================================
Below I will share some basic tips that you can use while trying to scale:
1. If you do not know which scaler to use, apply all and check the effect on the models.
2. If you do not understand the data, use standard scaler. It works most of the times.
3. If you know the max and min values of the feature, then use min max scaler. Like in CNN.
4. If most of the values in the feature column is 0 or sparce matrix, then use Max Absolute Scaling
5. If the data has outliers, use Robust Scaling.

```python
import numpy as np
df = pd.DataFrame(np.random.randint(-100,100,size=(100, 1)), columns=['Age'])
```

===================================================Standartization====================================================================

```python
from sklearn.preprocessing import StandardScaler

df_std_scaler = df.copy()
sc = StandardScaler() # creating an instance of the class object
df_new = pd.DataFrame(sc.fit_transform(df), columns=df.columns)  #fit and transforming StandardScaler the dataframe 
df_std_scaler['Age']= pd.DataFrame(sc.fit_transform(pd.DataFrame(df_std_scaler['Age'])), columns=['Age'])


df_std_scaler.head(10)
```

============================================Normalization==========================================================

Min Max scaling

```python
#From 0 to 1
from sklearn.preprocessing import MinMaxScaler

df_min_max = df.copy()
m1 = MinMaxScaler() # creating an instance of the class object
df_min_max['Age']= pd.DataFrame(m1.fit_transform(pd.DataFrame(df_min_max['Age'])), columns=['Age'])


df_min_max.head(10)
```

Mean Normalization

MaxAbsScaler

```python
#From -1 to 1
from sklearn.preprocessing import MaxAbsScaler

df_abs_scaler = df.copy()
m2 = MaxAbsScaler() # creating an instance of the class object
df_abs_scaler['Age']= pd.DataFrame(m2.fit_transform(pd.DataFrame(df_abs_scaler['Age'])), columns=['Age'])


df_abs_scaler.head(10)

```

Robust Scaler

```python
from sklearn.preprocessing import RobustScaler

df_robust_scaler = df.copy()
rs = RobustScaler() # creating an instance of the class object
df_robust_scaler['Age']= pd.DataFrame(rs.fit_transform(pd.DataFrame(df_robust_scaler['Age'])), columns=['Age'])
df_robust_scaler.head(10)
```
